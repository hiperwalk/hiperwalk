## Continuous-time quantum walk (CTQW) on a cycle

```python
import hiperwalk as hpw
import numpy as np
hb="y"
myHPC_option="cpu" if hb == "y" else None

hpw.set_hpc(myHPC_option)

#import os
#os.environ["OMP_NUM_THREADS"] = "2" # não altera o valor alterior atribuido ao executar o jupyter na linha de comando
import os;
print('num nucleos    =', os.cpu_count());
print('OMP_NUM_THREADS=', os.environ.get('OMP_NUM_THREADS'));
print('MKL_NUM_THREADS=', os.environ.get('MKL_NUM_THREADS'))
```

### Defining the graph and a QW instance

The next command creates a cycle, which is an instance of the Graph class.

```python
N = 101*10*2//2
cycle = hpw.Cycle(N)
```

The next command creates a QW on the cycle, which is an instance of the ContinuousTime class.

```python
import time
import sys
inicioG = time.perf_counter()
qw = hpw.ContinuousTime(cycle, gamma=1/(2*np.sqrt(2)), time=1)
fimG    = time.perf_counter()
print(f"cycle: Tempo decorrido: {fimG - inicioG:.6f} segundos", file=sys.stderr)
```

Note that the evolution operator is automatically created using 'time'=1, which coincides with the default value.

### Setting the initial state

```python
inicioK = time.perf_counter()
psi0 = qw.ket(N // 2)
fimK    = time.perf_counter()
print(f"ket: Tempo decorrido: {fimK - inicioK:.6f} segundos", file=sys.stderr)
```

### Probability distribution after $50$ steps  

```python
inicioK = time.perf_counter()
psi_final = qw.simulate(range_=(1,51), state=psi0)
fimK    = time.perf_counter()
print(f"ket: Tempo decorrido: {fimK - inicioK:.6f} segundos", file=sys.stderr)
prob = qw.probability_distribution(psi_final)
hpw.plot_probability_distribution(prob, plot='line', figsize=(6,3))
```

### Changing the labels

We can change the labels of the vertices using a Python dictionary.

```python
labels = {i: i - N // 2 for i in range(0, N, N//10)}
hpw.plot_probability_distribution(prob, plot='line',figsize=(6,3),labels=labels)
```

### Animation of the quantum walk

```python
list_of_states = qw.simulate(range = (5,50,2), state = psi0)
list_of_probs = qw.probability_distribution(list_of_states)
hpw.plot_probability_distribution(list_of_probs, range = (5,50,2), plot='line', animate=True, figsize=(7,3))
```
